```python
import pandas as pd
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
import pandas as pd
df = pd.DataFrame(boston["data"],columns=boston["feature_names"])
df["target"]=boston["target"]
df.to_csv("boston.csv",encoding="utf-8",index=False)
df
```

```python
from sklearn.model_selection import train_test_split
#train_test_split â†’ (ç‰¹å¾µ90%,ç‰¹å¾µ10%,ç›®æ¨™90%,ç›®æ¨™10%)
#[90% data, 10% data, 90% target, 10% target]
x_train,x_test, y_train, y_test=train_test_split(boston["data"],boston["target"],test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train,y_train)
```

```python
import graphviz
from sklearn.tree import  export_graphviz
g = export_graphviz(reg,out_file=None,feature_names = boston["feature_names"],filled=True)
graph = graphviz.Source(g)
graph
```

```python
pre = reg.predict(x_test)
print("é æ¸¬çµæžœ:",list(pre))
print("çœŸæ­£æ¨™ç±¤:",list(y_test))
from sklearn.metrics import r2_score
print("r2_score:",r2_score(y_test,pre))
```
