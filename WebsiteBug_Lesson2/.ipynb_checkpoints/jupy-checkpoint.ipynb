```python
#處理時間，秒數→格林威治時區
from datetime import datetime
str(datetime.utcfromtimestamp(1140825600))
```

```python
#轉換流程 → 遞給整行 → 每一個走過流程
def timetransfor (s):
    return str(datetime.utcfromtimestamp(s))
#apply:apply(流程名字) 對每一個執行流程
df["filmutc"]=df["film_date"].apply(timetransfor)
df
```

```python
#篩選操作：原理 df[[跟你的列一樣多的Boolean]]
df.head(3)[[True,False,True]]
```

```python
#篩選: 轉換操作→2550 True or False
#轉換流程 → 遞給整行 → 每一個走過流程
def filterflow (n):
    if n <= 5000 and n >= 3000:
        return True
    else:
        return False

#apply:apply(流程名字) 對每一個執行流程
df["comments"].apply(filterflow)
filter=df["comments"].apply(filterflow)
df[filter]
```

```python
#eval() → 把傳入的字串當成一個Python程式在執行
eval("print(3+2)")
a=eval("[1,2,3,4]")
a
```

```python
df["tags"] #字串(文字)

#eval() → 把傳入的字串當成一個Python程式在執行

def tagsflow(s):
    if "creativity" in eval(s):
        return True
    else:
        return False
    
df[df["tags"].apply(tagsflow)]
```
