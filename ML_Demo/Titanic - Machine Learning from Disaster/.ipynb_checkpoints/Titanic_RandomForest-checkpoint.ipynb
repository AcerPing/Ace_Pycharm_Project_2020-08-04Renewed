```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# row只顯示15個和column只顯示10個
# pd.set_option('display.max_rows', 15)
# pd.set_option('display.max_columns', 10)
```

```python
# 讀入鐵達尼訓練集
df = pd.read_csv('train.csv')
df
```

```python
# Step1: 遺漏值的處理
# 先檢查一下到底有那些欄位是有缺失值，而且缺失幾個。
df.isnull()
```

```python
df.isnull().sum()
```

```python
# 連續數值的缺失處理
# 補上median中位數值
df.median()
```

```python
# 注意'Pclass'是「種類」資料，試想若大部分乘客是做"經濟艙"，但中間乘客是做"頭等艙"...補缺失值當然是要選擇"經濟艙"。
med = df.median().drop('PassengerId').drop('Survived').drop('Pclass')
med
```

```python
df = df.fillna(med)
df.isnull().sum()
```

```python
# Cabin實在遺漏太多了(687/891)，所以直接選擇放棄這個欄位。
df = df.drop(['Cabin'], axis=1)
df
```

```python
# 類別(字串)的缺失處理
# 補上最常出現的字串
df['Embarked'].value_counts() # 'Embarked'補上最常出現的值
```

```python
Fillna_Embarked = df['Embarked'].value_counts().idxmax()
print(f'Embarked最常出現：{Fillna_Embarked}')
```

```python
df['Embarked'] = df['Embarked'].fillna(Fillna_Embarked)
df
```

```python
df.isnull().sum()
```

```python
# Step2: 特徵的處理
# One-Hot-Encoding
dummy = pd.get_dummies(df['Embarked'])
dummy
```

```python
df = pd.concat([df, dummy], axis=1) # concat是連結的意思，axis=1指的是水平的連接。
df = df.drop(['Embarked'], axis=1)
df
```

```python
dummy = pd.get_dummies(df['Sex'])
dummy
```

```python
df = pd.concat([df, dummy], axis=1)
df = df.drop(['Sex'], axis=1)
df
```

```python
# 從名字中取稱謂
# Petranec, Miss. Matilda
def transfer_Name(s):
    return s.split(',')[-1].split('.')[0].strip()

df['Name'] = df['Name'].apply(transfer_Name)
df
```

```python
# 看一下每個稱謂出現了幾次 → 發現Mr、Miss、Mrs是三個出現最多而且看起來真的有影響的稱謂，其餘都是一些稀少稱謂。
df['Name'].value_counts()
```

```python
NameList = df['Name'].value_counts()[0:3]
NameList
```

```python
def Name_Filter(name):
    if name in NameList: return name
    else: return 'Unknown'

df['Name'] = df['Name'].apply(Name_Filter)
df
```

```python
dummy = pd.get_dummies(df['Name'])
dummy
```

```python
df = pd.concat([df, dummy], axis=1)
df = df.drop(['Name', 'Unknown'], axis=1)
df
```

```python
# 丟棄Ticket及	PassengerId
df = df.drop(['Ticket', 'PassengerId'], axis=1)
df
```

```python
# 分成測試資料與訓練資料
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(df.drop(['Survived'], axis=1), df['Survived'], test_size=0.1)
```

```python
# 畫個heatmap → 性別為最重要的差別
plt.figure(figsize=(14, 10))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.astype(float).corr(), cmap='BrBG', linewidths=0.1, square=True, linecolor='white', annot=True)
```

```python
# 開始預測
# 先使用一個分類樹看看會怎麼處理問題
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=15, min_samples_leaf=20)
clf = clf.fit(data_train, target_train)
clf
```

```python
from sklearn.tree import export_graphviz
import graphviz
dot_data = export_graphviz(clf, out_file=None, feature_names=df.drop(['Survived'], axis=1).columns, class_names=['No', 'Yes'], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render('Titanic_RandomForest') # 輸出pdf
graph
```

```python
from sklearn.metrics import accuracy_score
predict = clf.predict(data_test)
print(f'預測: {predict}')
print(f'正確標籤: {target_test.to_list()}')
print(f'正確率: {accuracy_score(target_test, predict)*100}%')
```

```python
# 建立模型
# 使用隨機森林來建立模型，不過這次不使用隨機切割(train_test_split)，而是做交叉驗證(cross_val_score)
from sklearn.ensemble import RandomForestClassifier
# 參數說明
# n_estimators: 要有幾棵樹
# max_depth: 每棵樹的深度
clf = RandomForestClassifier(n_estimators=26, max_depth=6)
clf
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
# cv參數代表要做幾次的交叉驗證
scores = cross_val_score(clf, df.drop(['Survived'], axis=1), df['Survived'], cv=10, n_jobs=-1) # 不用fit了，fit和predict會由交叉驗證幫忙做
print(f'十次分數: {scores}')
# 最後算出每次驗證的平均
print(f'平均: {np.average(scores)}') # 由於是ndarray，可以直接使用average來計算平均。
```

# 〔補充〕GridSearchCV 貪婪演算法，找出最佳參數。 #

```python
# GridSearchCV貪婪演算法，找出最佳參數。
from sklearn.model_selection import GridSearchCV
```

```python
# DecisionTreeClassifier
clf = DecisionTreeClassifier()
params = {'max_depth':range(5,20), 'min_samples_leaf':range(1,30)}
search = GridSearchCV(clf, params, cv=10, n_jobs=-1) # n_jobs: 平行運算，一次執行幾份工作。
search.fit(data_train, target_train) # np.array(date_train), target_train
search
```

```python
print(f'最佳參數: {search.best_params_}')
print(f'最佳分數: {search.best_score_}')
```

```python
# RandomForestClassifier
clf = RandomForestClassifier()
params = {'max_depth': range(1,10), 'n_estimators':range(20,50)}
search = GridSearchCV(clf, params, cv=10, n_jobs=-1)
search.fit(df.drop(['Survived'], axis=1), df['Survived'])
search
```

```python
print(f'最佳參數:{search.best_params_}')
print(f'最佳分數:{search.best_score_}')
```
