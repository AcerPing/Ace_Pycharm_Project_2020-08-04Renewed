```python
import pandas as pd
df_train = pd.read_csv('train.csv', encoding='utf-8')
df_train
```

```python
df_test = pd.read_csv('test.csv', encoding='utf-8')
df_test
```

```python
# (訓練資料)有哪些缺失值
df_train.isna()
```

```python
# (訓練資料)各欄位有多少缺失值
df_train.isna().sum()
```

```python
# (測試資料)有哪些缺失值
df_test.isna()
```

```python
# (測試資料)各欄位有多少缺失值
df_test.isna().sum()
```

## Step1.1 補缺值：數值（中位數）

```python
# (訓練資料)若欄位為數值型態，則缺失值補上中位數
med = df_train.median() # 中位數數值
med
```

```python
# (訓練資料)若欄位為數值型態，則缺失值補上中位數
df_train = df_train.fillna(med) 

# PassengerId → 沒差
# Pclass → 因Test測試資料沒缺，所以沒差

df_train
```

```python
# (訓練資料)各數值欄位補上中位數後，還有多少缺失值
df_train.isna().sum()
```

```python
# (測試資料)若欄位為數值型態，則缺失值補上中位數，但中位數數值以訓練資料計算時的為主。
# 千萬不要再重新算一次中位數
df_test = df_test.fillna(med) 
```

```python
df_test.isna().sum()
```

## Step1.2 補缺值：類別（最常出現的值）

```python
# Cabin 船艙號碼
# 1. 每個人船艙號碼相異且不重覆
# 2. 太多缺失值
# 所以直接Drop這個欄位
```

```python
# (訓練資料) 計算Embarked欄位各值出現的次數
df_train['Embarked'].value_counts()
```

```python
# (訓練資料)Embarked欄位的缺失值補上次數出現最多的值
most = df_train['Embarked'].value_counts().idxmax()
df_train['Embarked'] = df_train['Embarked'].fillna(most)
```

```python
# (訓練資料)還有多少缺失值
df_train.isna().sum()
```

```python
# (測試資料)補上'Embarked'缺失值
# 千萬不要再重新計算
df_test['Embarked'] = df_test['Embarked'].fillna(most)
df_test.isna().sum()
```

## Step2 類別 → One-Hot-Encoding

```python
# 可轉可不轉的欄位：
# 1. Pclass → 可以承認有大小關係
# 2. Sex → 就只有兩種值 
```

```python
# Convert categorical variable into dummy/indicator variables.
df_train = pd.get_dummies(df_train, columns=['Sex', 'Embarked'])

# 參數
# columns: Column names in the DataFrame to be encoded. If columns is None then all the columns with object or category dtype will be converted.

df_train # 轉換完後，原本的Column會被取代。 
```

```python
df_test = pd.get_dummies(df_test, columns=['Sex', 'Embarked'])
df_test
```

```python
# 從欄位'Name'取出稱謂
# 試水溫，看看一格怎麼做
s = 'Wirz, Mr. Albert'
s = s.split(',')[-1].split('.')[0].strip()
```

```python
# 看一下多寡
def name_split(s):
    s = s.split(',')[-1].split('.')[0].strip()
    return s
    
df_train["Name"].apply(name_split).value_counts()
```

```python
# 看一下多寡  
df_test["Name"].apply(name_split).value_counts()
```

```python
def name(s):
    reserved = ['Mr', 'Miss', 'Mrs', 'Master']
    s = s.split(',')[-1].split('.')[0].strip()
    if s in reserved: return s
    else: return None
    
df_train["Name"] = df_train["Name"].apply(name)
df_train
```

```python
df_train = pd.get_dummies(df_train, columns=['Name'])
df_train
```

```python
df_test["Name"] = df_test["Name"].apply(name)
df_test
```

```python
df_test = pd.get_dummies(df_test, columns=['Name'])
df_test
```

```python
# Drop specified labels from rows or columns.
df_train = df_train.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)
df_train
```

```python
# 測試資料一定要保留PassengerId，才能交答案。
Test_PassengerId = df_test['PassengerId']
df_test = df_test.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)
df_test
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df_train.corr() # 計算相關係數
```

```python
# 熱度圖
plt.figure(figsize=(14, 14)) # 調整圖像大小，單位inch
sns.heatmap(df_train.corr(), annot=True, cmap='RdBu_r')
# annot:顯示相關係數數值；If True, write the data value in each cell. If an array-like with the same shape as data, then use this to annotate the heatmap instead of the data. Note that DataFrames will match on position, not index. → bool
# cmap:調整顏色；The mapping from data values to color space. If not provided, the default will depend on whether center is set. → matplotlib colormap name

# matplotlib colormap (Diverging): https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
y_train = df_train['Survived'] # 訓練資料的答案
x_train = df_train.drop(['Survived'], axis = 1) # 訓練資料的題目
x_predict = df_test
```

```python
# 粗略調整 # GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier() #此處先不帶參數，之後由GridSearchCV帶入

# 要調整的每個參數及範圍
params = {
    'n_estimators':range(25, 35),
    'max_depth':range(5, 10)
}

grid = GridSearchCV(clf, params, n_jobs=-1, cv=10)
# grid = GridSearchCV(clf, params, n_jobs=4, cv=10)
# estimator: classifier決策樹分類器；This is assumed to implement the scikit-learn estimator interface. → estimator object
# param_grid: 參數範圍(n_estimators、max_depth)；Dictionary with parameters names (str) as keys and lists of parameter settings to try as values, or a list of such dictionaries, in which case the grids spanned by each dictionary in the list are explored. This enables searching over any sequence of parameter settings. → dict or list of dictionaries
# n_jobs: 分成數個工作同時間執行；Number of jobs to run in parallel. → int

# 採用隨機森林，每次隨機放棄一些訓練資料，所以每次出來的結果會不太一樣。
grid.fit(x_train, y_train)
print('最佳參數:', grid.best_estimator_)
# best_estimator_: Parameter setting that gave the best results on the hold out data. → dict
print('最佳分數:', grid.best_score_)
# best_score_: Mean cross-validated score of the best_estimator → float
```

```python
# 細微調整
from sklearn.model_selection import cross_val_score
import numpy as np

clf = RandomForestClassifier(n_estimators=27, max_depth=7)
# n_estimators:幾個分類器(決策樹)；The number of trees in the forest. → int
# max_depth: 層樹，每個決策樹有幾層；The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. → int

score = cross_val_score(clf, x_train, y_train, cv=10)
# estimator:決策樹；The object to use to fit the data.
# X:訓練資料；The data to fit. Can be for example a list, or an array.
# y:答案；The target variable to try to predict in the case of supervised learning.
# cv:要切幾分；Determines the cross-validation splitting strategy. → int
print(f'十次分數:{score}')
print(f'十次平均:{np.average(score)}')
```

```python
# 確定參數: 27, 7
clf = RandomForestClassifier(n_estimators=27, max_depth=7)
clf.fit(x_train, y_train)
pre = clf.predict(x_predict)
pre
```

```python
result = pd.DataFrame(
    {'PassengerId':Test_PassengerId,
    'Survived': pre},
    columns=['PassengerId', 'Survived']
)

result
```

```python
result.to_csv('Titanic_RandomForest.csv', encoding='utf-8', index=False)
```

```python
# 觀察每個特徵重要性
pd.DataFrame({
    'Name': x_train.columns,
    'Importance': clf.feature_importances_},
    columns=['Name', 'Importance'])
```

### 測試改用其它演算法（KNN）的命中機率

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_predict_scale = scaler.transform(x_predict)
pd.DataFrame(x_train_scale)

# fit: 找出最大值與最小值；Compute the minimum and maximum to be used for later scaling.
# fit_transform: 依照最大值與最小值差距，將數值轉換為0~1之間；Fit to data, then transform it.
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier() #此處先不帶參數，之後由GridSearchCV帶入

# 要調整的每個參數及範圍
# n_neighbors: 要找幾個鄰居；Number of neighbors to use by default for kneighbors queries.
params = {
    'n_neighbors':range(5, 100),
}

grid = GridSearchCV(clf, params, n_jobs=-1, cv=10)
# estimator: 分類器；This is assumed to implement the scikit-learn estimator interface. → estimator object
# param_grid: 參數範圍；Dictionary with parameters names (str) as keys and lists of parameter settings to try as values, or a list of such dictionaries, in which case the grids spanned by each dictionary in the list are explored. This enables searching over any sequence of parameter settings. → dict or list of dictionaries
# n_jobs: 分成數個工作同時間執行；Number of jobs to run in parallel. → int

# 此處不需要多次重跑，因為Cross Validation每次切割成的十份皆為固定，所以每次結果皆一樣。
grid.fit(x_train_scale, y_train) #注意:要用Scale過後的結果(x_train_scale)
print('最佳參數:', grid.best_estimator_)
# best_estimator_: Parameter setting that gave the best results on the hold out data. → dict
print('最佳分數:', grid.best_score_)
# best_score_: Mean cross-validated score of the best_estimator → float
```

```python
clf = KNeighborsClassifier(n_neighbors=22)
clf.fit(x_train_scale, y_train)
pre = clf.predict(x_predict_scale)


result = pd.DataFrame(
    {'PassengerId':Test_PassengerId,
    'Survived': pre},
    columns=['PassengerId', 'Survived']
)

result
```

```python
result.to_csv('Titanic_KNN.csv', encoding='utf-8', index=False)
```
