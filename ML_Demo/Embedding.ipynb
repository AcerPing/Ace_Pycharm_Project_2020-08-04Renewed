<H2>Step1. 資料預處理</H2>

```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_rows', 15)
pd.set_option('display.max_columns', 10)
train_df = pd.DataFrame(columns=['content', 'sentiment'])
train_df 
```

```python
# 走過train\pos的資料夾
pos_path = r'D:\Python_Summarize\Ace_Pycharm_Project\Ace_Deep_Learning\aclImdb\train\pos'
for fpath in os.listdir(pos_path):
    if not fpath.startswith('.'):
        fpath = os.path.join(pos_path, fpath)
        f = open(fpath, 'r', encoding='utf-8')
        content = f.read()
        s = pd.Series([content, 1], index=['content', 'sentiment'])
        train_df = train_df.append(s, ignore_index=True)

train_df
```

```python
# 走過train\neg的資料夾
neg_path = r'D:\Python_Summarize\Ace_Pycharm_Project\Ace_Deep_Learning\aclImdb\train\neg'
for fpath in os.listdir(neg_path):
    if not fpath.startswith('.'):
        fpath = os.path.join(neg_path, fpath)
        f = open(fpath, 'r', encoding='utf-8')
        content = f.read()
        s = pd.Series([content, 0], index=['content', 'sentiment'])
        train_df = train_df.append(s, ignore_index=True)
        
train_df
```

```python
test_df = pd.DataFrame(columns=['content', 'sentiment'])
test_df 
```

```python
# 走過test\pos的資料夾
pos_path = r'D:\Python_Summarize\Ace_Pycharm_Project\Ace_Deep_Learning\aclImdb\test\pos'
for fpath in os.listdir(pos_path):
    if not fpath.startswith('.'):
        fpath = os.path.join(pos_path, fpath)
        f = open(fpath, 'r', encoding='utf-8')
        content = f.read()
        s = pd.Series([content, 1], index=['content', 'sentiment'])
        test_df = test_df.append(s, ignore_index=True)

test_df
```

```python
# 走過test\neg的資料夾
neg_path = r'D:\Python_Summarize\Ace_Pycharm_Project\Ace_Deep_Learning\aclImdb\test\neg'
for fpath in os.listdir(neg_path):
    if not fpath.startswith('.'):
        fpath = os.path.join(neg_path, fpath)
        f = open(fpath, 'r', encoding='utf-8')
        content = f.read()
        s = pd.Series([content, 0], index=['content', 'sentiment'])
        test_df = test_df.append(s, ignore_index=True)

test_df
```

```python
# 把訓練資料找出最常見的2000字拿來建立出2000維度的字典
from keras.preprocessing.text import Tokenizer
token = Tokenizer(num_words=2000)
token.fit_on_texts(train_df['content'])
token.word_index
```

```python
# 利用字典把我們整理好的DataFrame每一個詞變成數字
x_train_seq = token.texts_to_sequences(train_df['content'])
pd.DataFrame(x_train_seq)
```

```python
x_test_seq = token.texts_to_sequences(test_df['content'])
pd.DataFrame(x_test_seq)
```

```python
# 每一篇文章的長度不一樣，截長補短把每一篇文章取100個字來做訓練
# 訓練的時候無法處理長短不同的文章
from keras.preprocessing import sequence
x_train_pad = sequence.pad_sequences(x_train_seq, maxlen=100)
pd.DataFrame(x_train_pad)
```

```python
x_test_pad = sequence.pad_sequences(x_test_seq, maxlen=100)
pd.DataFrame(x_test_pad)
```

<H2>Step2. 建立模型</H2>

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Flatten
from keras.layers.embeddings import Embedding

model = Sequential()
model.add(Embedding(output_dim=32, input_dim=2000, input_length=100)) # Keras內建詞嵌入層，把2000維度的每一個詞向量換成32維度的詞向量
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))
```

```python
model.add(Dense(units=1, activation='sigmoid')) #因為我們是二元分類,最後的激勵函數選擇sigmoid 
# sigmoid(正+負=100%)；softmax(類別全部=100%)
```

```python
model.summary()
```

<H4>Param參數個數</H4>
<p>Embedding層：2000*32=64,000</p>
<p>Flatten層：32（詞維度）*100（文章的字數）=3,200</p>

```python
import numpy as np
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # 因為只是二元分類, 所以loss選擇binary_crossentropy
```

```python
from keras.datasets import imdb
y_train = train_df['sentiment'].astype('float32')
train_history = model.fit(x_train_pad, y_train, batch_size=100, epochs=3, verbose=2, validation_split=0.2)
```

```python
y_test = test_df['sentiment'].astype('float32')
model.evaluate(x_test_pad, y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(train_history.history['loss'])
plt.plot(train_history.history['val_loss'])
plt.title('Loss Graph')
plt.legend(['loss', 'val_loss'], loc='upper left')
```

<H2>Step3. RNN</H2>

```python
from keras.layers import SimpleRNN
model = Sequential()
model.add(Embedding(output_dim=32, input_dim=2000, input_length=100))
model.add(Dropout(0.2))

model.add(SimpleRNN(units=16)) # RNN:記憶16個狀態
model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.35))

model.add(Dense(units=1, activation='sigmoid')) # 因為是二元分類,最後的激勵函數選擇sigmoid

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #二元分類,所以這裡的loss選擇binary_crossentropy

model.summary()
```

```python
train_history = model.fit(x_train_pad, y_train, batch_size=100, epochs=3, verbose=2, validation_split=0.2)
y_test = test_df['sentiment'].astype('float32')
model.evaluate(x_test_pad, y_test) # 正確率是list第二個元素
```
