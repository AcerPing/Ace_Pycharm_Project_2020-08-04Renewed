```python
from keras.datasets.mnist import load_data # 載入訓練資料
load_data() #結構：Type→Tuple；((訓練資料), (測試資料))；((X_Train, Y_Train), (X_Test, Y_Test))
((X_Train, Y_Train), (X_Test, Y_Test)) = load_data()
```

```python
# .shpae → 維度
print(X_Train.shape) #6,0000筆訓練資料、圖片像素尺寸:28*28
print(Y_Train.shape)
print(X_Test.shape) #1,0000筆測試資料、圖片像素尺寸:28*28
print(Y_Test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xticks([])
plt.yticks([])
plt.title('This is ' + str(Y_Train[0])) # 標題(答案)
plt.imshow(X_Train[0], cmap='gray') # mnist為黑白圖
```

```python
# 轉成表格
import pandas as pd
pd.DataFrame(X_Train[0])
```

```python
# 全連接層: Fully-Connected (Dense Layer)
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
```

```python
# units: 有多少個神經元；Positive integer, dimensionality of the output space. 
# activation: 激活函式；Activation function to use. If you don't specify anything, no activation is applied
# Input shape (input_dim): 在第一層的時候要鍵入輸入的圖片尺寸(EX. 28*28=764)
```

```python
model.add(Dense(256, activation='relu', input_dim=784)) # 第一層；參數計算=784*256(Dense)+256(bias)=200960
```

```python
model.add(Dense(10, activation='softmax')) # 最後一層；參數計算=256*10(Dense)+10(bias)=2570
```

```python
model.summary()
```

```python
# 預處理資料(題目)
# 1.) MLP輸入: 1D，一維資料
# 2.) 配合框架隨機的Weight: 所有輸入資料要Normalize(-1~1 或 0~1)

# reshape為60,000筆資料、每筆資料維度為784 (60000, 784)後，Normalize(除上255)。
X_Train_reshape = X_Train.reshape(-1, 784) / 255  
X_Test_reshape = X_Test.reshape(-1, 784) / 255  
```

```python
print(X_Train_reshape.shape)
X_Train_reshape
```

```python
print(X_Test_reshape.shape)
X_Test_reshape
```

```python
# 預處理資料(答案)
# 假若答案為0，則0-9的機率: [1, 0, 0, 0, ....] (經過Softmax的十個機率)
# 假若答案為1，則0-9的機率: [0, 1, 0, 0, ....] (經過Softmax的十個機率)
# →→→ One-Hot-Encoding
from keras.utils.np_utils import to_categorical
Y_Train_Cat = to_categorical(Y_Train)
Y_Test_Cat = to_categorical(Y_Test)
```

```python
Y_Train_Cat
```

```python
Y_Test_Cat
```

```python
print(Y_Train[0]) # 未轉化前
print(Y_Train_Cat[0]) # 轉化後
```

```python
# compile: 確定模型的訓練方式
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy']) 
# loss: 預測與正確答案的差距
# accuracy: 預測準確率
```

```python
# fit：訓練模型

# batch_size：看多少筆再做梯度的調整
# 1 → 每看一筆就調整權重weights；缺點：亂走
# 60,000 → 看完全部再調整一次；缺點：花時間
# 建議數值：看完100~200筆平均調整一次weights

# 〈重要！！！〉 validation_split：用沒看過的資料（測試資料），檢查模型什麼時候最好！最好的時候停止訓練。（train_test_split）

# epochs：整份訓練資料要看幾輪

# validation + epochs → 控制擬合程度
# 當validation表現最好的時候，就停下來。

# verbose：log顯示模式；Verbosity mode. 0 = silent, 1 = progress bar, 2 = one line per epoch.

# 10 epochs、200 batch_size、0.1 validation_split → 54000*10/200=2700次梯度更新
model.fit(X_Train_reshape, Y_Train_Cat,
         batch_size=200,
         validation_split=0.1,
         epochs=15,
         verbose=2)
```

```python
# 衡量模型（方法一）
model.evaluate(X_Test_reshape, Y_Test_Cat)
```

```python
# 預測
# predict → 預測機率
# predict_classes → 預測類別(答案)
pre = model.predict_classes(X_Test_reshape)
pre
```

```python
# 衡量模型（方法二－Confusion Matrix）
from sklearn.metrics import confusion_matrix
index_i = ['{} (正確)'.format(i) for i in range(10)]
column_c = ['{} (預測)'.format(i) for i in range(10)]
pd.DataFrame(confusion_matrix(Y_Test, pre),
            index=index_i,
            columns=column_c)
```

```python
import numpy as np
# nonzero() → type為tuple，(1st索引, 2nd索引, 3rd索引...)
idx = np.nonzero(pre != Y_Test)[0]
print('預測錯誤：' + str(len(idx)))

pre_false_img = X_Test[idx]
pre_false_label = Y_Test[idx]
pre_false_predict = pre[idx]

width = 10
height = len(idx) // width + 1
plt.figure(figsize=(14, 42)) # 寬、高

for i, img in enumerate(pre_false_img):
    plt.subplot(height, width, i+1)
    t = '[Pre]: {}\n[Ori]: {}'.format(pre_false_predict[i], pre_false_label[i])
    plt.title(t)
    plt.axis('off')
    plt.imshow(img, cmap='gray')
```
