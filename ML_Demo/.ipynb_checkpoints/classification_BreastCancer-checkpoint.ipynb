```python
# 1. 載入資料集
from sklearn.datasets import load_breast_cancer
import pandas as pd
BreastCancer = load_breast_cancer()
BreastCancer
```

```python
# 2. 利用pandas建立表格
df = pd.DataFrame(BreastCancer.data, columns=BreastCancer.feature_names) #Samples total: 569
df['target'] = BreastCancer.target
df
```

```python
# 匯出excel及csv
df.to_excel('BreastCancer.xlsx', sheet_name='BreastCancer', index=False)
df.to_csv('BreastCancer.csv')
```

```python
# 拆分資料：分為 測試資料 及 訓練資料
# 指定：90%為訓練資料，10%為測試資料
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(BreastCancer.data, BreastCancer.target, test_size=0.1)
```

```python
# 開始模型訓練 fit
# 指定criterion為entropy
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy', max_depth=4)

# 參數說明
# criterion：The function to measure the quality of a split. Supported criteria are “gini”.

clf
```

```python
clf = clf.fit(X_train, y_train)
clf
```

```python
# 預測
predict_test = clf.predict(X_test)
print(f'正確資料：{y_test}')
print(f'預測資料：{predict_test}')
```

```python
# 得分
from sklearn.metrics import accuracy_score
point = accuracy_score(y_test, predict_test)*100
print(f'預測分數：{point}%')
```

```python
# 畫圖-1
from sklearn.tree import export_graphviz
import graphviz
graph = export_graphviz(clf)
graphviz.Source(graph)
```

```python
# 畫圖-2
graph = export_graphviz(clf, feature_names=BreastCancer.feature_names, class_names=BreastCancer.target_names, filled=True)
graphviz.Source(graph)
```

```python
# 利用混淆矩陣查看那些算錯
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, predict_test))
```
