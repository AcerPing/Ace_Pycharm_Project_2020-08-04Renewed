```python
from sklearn.datasets import load_wine # 載入資料集(sklearn wine)
wine = load_wine()
wine
```

```python
import pandas as pd
df = pd.DataFrame(wine['data'], columns=wine.feature_names)
df
```

## 狀況一：已知有三種分類 ##

```python
from sklearn.cluster import KMeans
cluster_wine = KMeans(n_clusters=3) # 分成三類
cluster_wine.fit(wine['data']) #訓練資料
cluster_wine.labels_ # 標籤
```

```python
df['labels'] = cluster_wine.labels_
df
```

```python
cluster_wine.cluster_centers_ # 圓心
```

```python
# 畫圖
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# heatmap熱圖
plt.figure(figsize=(25,20))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.astype(float).corr(), cmap='BrBG',\
           linewidths=0.1, square=True, linecolor='white',\
           annot=True)
```

```python
# scatter plot散點圖
sns.scatterplot(x=df['proline'], y=df['magnesium'], hue=df['labels'])
```

## 狀況二：不知到有幾種分類（未知K） ##
### 把所有K試過一遍，對每個K，我可以求「平均信心程度」，選一個最高的。 ###

```python
from sklearn.metrics import silhouette_score

scores, ks = list(), list()

for k in range(2, 11):
    # k值不得為1。ValueError: Number of labels is 1. Valid values are 2 to n_samples - 1 (inclusive)
    # Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.
    cluster_wine = KMeans(n_clusters=k) # 分成k類
    cluster_wine.fit(wine['data']) #訓練資料
    score = silhouette_score(wine['data'], cluster_wine.labels_)
    scores.append(score)
    ks.append(k)

print("K值：", ks)
print("親疏有別的分數：", scores)
```

```python
df_silhouette_score = pd.DataFrame({'silhouette_score': scores}, index=ks)
df_silhouette_score
```

```python
plt.title("silhouette_score")
plt.xlabel("K")
plt.ylabel("score")
plt.plot(ks, scores,  'o--m') # 折線圖
```

```python
# 依照silhouette_score由大到小排序
df_silhouette_score.sort_values(by='silhouette_score', ascending=False)
```
